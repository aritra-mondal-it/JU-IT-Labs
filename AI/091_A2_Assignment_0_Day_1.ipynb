```python
# Task 3 - Perform normal dfs operation and showing all the intermediate steps.
```

```python
# Matrix DFS traversal with intermediate steps

# Direction vectors: left, down, right, up
dRow = [0, 1, 0, -1]
dCol = [-1, 0, 1, 0]

def is_valid(r, c, ROW, COL, visited):
    return (0 <= r < ROW and 0 <= c < COL and not visited[r][c])

def dfs_matrix(grid, start_r=0, start_c=0):
    ROW, COL = len(grid), len(grid[0])
    visited = [[False]*COL for _ in range(ROW)]
    stack = [(start_r, start_c)]

    while stack:
        r, c = stack.pop()

        if not is_valid(r, c, ROW, COL, visited):
            continue

        visited[r][c] = True
        print(f"Visiting cell ({r}, {c}) → {grid[r][c]}")

        # Push neighbors onto stack
        for i in range(4):
            nr, nc = r + dRow[i], c + dCol[i]
            if is_valid(nr, nc, ROW, COL, visited):
                stack.append((nr, nc))

# Example usage:
grid = [
    [ 1,  2,  3],
    [ 4,  5,  6],
    [ 7,  8,  9]
]
dfs_matrix(grid)
```

```python
import numpy as np

def get_matrix_input(name):
    """Prompts the user to input a 3x3 matrix."""
    print(f"Enter the elements of the {name} matrix (9 integers):")
    matrix = []
    for i in range(3):
        row = list(map(int, input(f"Row {i + 1}: ").split()))
        matrix.append(row)
    return np.array(matrix)

def is_valid_move(matrix, x, y):
    """Checks if the move is within the matrix bounds."""
    return 0 <= x < 3 and 0 <= y < 3

def iterative_dfs(start, target):
    """Performs iterative DFS to find the transformation path from start to target."""
    stack = [(start, [], set())]  # (current matrix, path, visited states)
    step = 0

    while stack:
        current, path, visited = stack.pop()
        visited.add(tuple(current.flatten()))

        print(f"Stack at Step {step}:")
        for state in stack:
            print(state[0])  # Print only the current matrix of each state
        print("-----")

        if np.array_equal(current, target):
            path.append((step, current.copy()))
            return path

        x, y = np.where(current == 0)
        x, y = x[0], y[0]  # Extract the row and column of the empty cell

        # Define possible moves: up, down, left, right
        moves = [(-1, 0), (1, 0), (0, -1), (0, 1)]
        for dx, dy in moves:
            nx, ny = x + dx, y + dy
            if is_valid_move(current, nx, ny):
                new_matrix = current.copy()
                new_matrix[x, y], new_matrix[nx, ny] = new_matrix[nx, ny], new_matrix[x, y]
                if tuple(new_matrix.flatten()) not in visited:
                    stack.append((new_matrix, path + [(step, current.copy())], visited))
        step += 1

    return None

def main():
    """Main function to execute the matrix transformation."""
    start = get_matrix_input("start")
    target = get_matrix_input("target")

    path = iterative_dfs(start, target)
    if path:
        print("\nTransformation path:")
        for step, matrix in path:
            print(f"Step {step}:")
            print(matrix)
    else:
        print("No transformation path found.")

if __name__ == "__main__":
    main()
```
